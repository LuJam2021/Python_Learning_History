# Backtesting 投資策略回測報酬率

backtrader => 進階

Trading View



```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover

from backtesting.test import SMA, GOOG #GOOG讀進去的data

#移動平均 繼承Strategy
class SmaCross(Strategy):
    def init(self):#沒底線=自定義初始化
        price = self.data.Close #價格=傳進去的Close價
        self.ma1 = self.I(SMA, price, 10) #10日移動平均線
        self.ma2 = self.I(SMA, price, 20) #20日移動平均線

    def next(self):
        if crossover(self.ma1, self.ma2):  #10日線穿過20日線(往下) 買進
            self.buy()
        elif crossover(self.ma2, self.ma1): #20日線穿過10日線(往上) 賣出
            self.sell()
```

```python
bt = Backtest(GOOG, SmaCross, commission=.002, exclusive_orders=True) #exclusive_orders=維持一筆交易
```

```python
stats = bt.run()
```
